```python
#!/usr/bin/env python3.10
# -*- coding: utf-8 -*-
"""
Created on  11.09.2023
@author: Paul
"""

import numpy as np
import matplotlib.pyplot as plt


d = 2   # number of states per site
nq = 2   # number of sites
D = 2   # bond dimension 

# This small programm solves for the eigenvalues of A 
# using a tensor train representation and variational approaches 
# in the framework of DMRG  - Refer to Nathalia Chepiga Slides of ETN SummerSchool 23


A =  np.diag([1,2,3,4]) # Define a Matrix
print(A.shape)

print(d*np.ones(nq*d, dtype=int))

# Reshape 
C_ijkl =  np.reshape(A, d*np.ones(nq*d, dtype=int))

C_ijkl =  np.transpose(C_ijkl,[0,2,1,3])

A_tilde = np.reshape(C_ijkl, A.shape)

# First decomposition to get first MPO term.
U, S, V_dagger = np.linalg.svd(A_tilde,full_matrices=False)
S = np.diag(S)
MPO1 = np.reshape(U[:,0:2], d*np.ones(nq+1,dtype=int)) # TODO: implement propper reshape vec
                  
print("MPO_shape", MPO1.shape)
print("MPO1", MPO1)

MPO2=  np.reshape(np.matmul(S,V_dagger)[0:2,:],d*np.ones(nq+1,dtype=int))

print("MPO_shape", MPO2.shape)
print("MPO2", MPO2)



# TEST
C_ijkl_test = np.zeros_like(C_ijkl)
for v1 in range(0,2):
    for v2 in range(0,2):
        C_ijkl_test[v1,v2] =  np.matmul(MPO1[:,v1,:], MPO2[:,:,v2])

#print("C_ijkl:", np.reshape(A, d*np.ones(nq*d, dtype=int)))
#print("C_ijkl_test:",C_ijkl_test)


print(np.reshape(C_ijkl_test,A.shape))
```
