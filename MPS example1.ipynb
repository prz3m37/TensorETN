```python
#!/usr/bin/env python3
import numpy as np

#THE Schollwöck way!

p = 2  #Number of states per site, qubit p = 2
N = 3  #Number of Sites  -  num qubits  Schollwöck -> L
D1 = 2 #Bond dimension
D2 = 2
D3 = 2
if p ==2 : 
    ket0 = np.array([[1],[0]])
    ket1 = np.array([[0],[1]])


shaper = np.ones(N,dtype=np.int8)*p

psi = np.zeros(p**N) #arbitary state vector
psi[7] = 1
psi[1] = 1
psi[0] = 1 
print("psi", psi)
print("##############")

################## LEFT CANONICAL FORM################
# GOING FROM i = 1...N
# create an array which should be decomposed 
C_ijkl = np.reshape(psi,shaper) # Tensor of size p x....x p
print("Cijkl:\n", C_ijkl)

print("##############")
#First Decomposition ########################### i = 1 
Psib = np.reshape(C_ijkl,[p,p**(N-1)])  # Mode 1 matricization of C_ijkl
print("Psib:\n", Psib)

#D1 = np.minimum(D1, np.linalg.matrix_rank(Psib))  Use adaptive compression here 
U1, S, V_dagger = np.linalg.svd(Psib,full_matrices=False)
S = np.diag(S)
print("Ushape:",U1.shape,"Sshape:", S.shape,"V_dagshape:", V_dagger.shape)
A1 = U1# Collection of D1 row vectors A_sigma1 in sigma_1 x alpha_1
print("A1: \n",A1)
#U1_approx  =  U1[:,:D1] # Compression by MPS   
print("S:\n",S) #Vector of sigular values
#S_approx = S[:D1,:D1]
print("V_dag:\n",V_dagger) # Matrix of right eigenvectors tau alpha2
#V_dagger_approx = V_dagger[:D1,:]

print("##############")
#Second Decomposition ########################### i = 2
Psib = np.reshape(np.matmul(S,V_dagger),[D2*p,p**(N-2)])
print("Psib:\n", Psib)
#D2 = np.minimum(D2, np.linalg.matrix_rank(Psib))  Use adaptive compression here 
U2, S, V_dagger = np.linalg.svd(Psib,full_matrices=False)
S = np.diag(S)
print("Ushape:",U2.shape,"Sshape:", S.shape,"V_dagshape:", V_dagger.shape)
print("U:\n",U2) #Matrix of left eigenvector tau alpha1
A2 = np.reshape(U2,[p,D1,D2]) #  collection of row vecs  sigma_2 x alpha_1 x alpha_2
print("A2:\n", A2)
print("S:\n",S) #Vector of sigular values
print("V:\n",V_dagger) # Matrix of right eigenvectors tau alpha2


print("##############")
#Last Decomposition ########################### i = N
Psib = np.reshape(np.matmul(S,V_dagger),[D3*p,p**(N-3)])
print("Psib:\n", Psib)
#D2 = np.minimum(D2, np.linalg.matrix_rank(Psib))  Use adaptive compression here 
U3, S, V_dagger = np.linalg.svd(Psib,full_matrices=False)
S = np.diag(S)
print("Ushape:",U3.shape,"Sshape:", S.shape,"V_dagshape:", V_dagger.shape)
print("U:\n",U3) #Matrix of left eigenvector tau alpha1
A3 = np.reshape(U3,[p,D2]) #  collection of row vecs  sigma_2 x alpha_1 x alpha_2
print("A3:\n", A3)
print("S:\n",S) #Vector of sigular values
print("V:\n",V_dagger) # Matrix of right eigenvectors tau alpha2


print("##########TEST##########")
C_ijkl_test = np.zeros_like(C_ijkl)
for sigma1 in range(0,D1):
    for sigma2 in range(0,D2):
        for sigma3 in range(0,D3):
            C_ijkl_test[sigma1,sigma2,sigma3] = np.matmul(np.matmul(A1[sigma1,:],A2[:,sigma2,:]),A3[:,sigma3])


print("C_IJKL:\n",C_ijkl)
print("C_IJKL_test:\n",C_ijkl_test)
```
